```python
%matplotlib ipympl
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
#suppresses weird generator output
from contextlib import contextmanager
import sys, os

@contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        old_stdout = sys.stdout
        sys.stdout = devnull
        try:  
            yield
        finally:
            sys.stdout = old_stdout


#plots p orbitals
with suppress_stdout():
    plt.style.use('default')
    fig = plt.figure(figsize=(12, 9), dpi=80)
    ax = fig.add_subplot(1, 3, 1, projection='3d')
    ax.set_box_aspect((2, 1, 1))
   
    u = np.linspace(-2.626, 2.626, 60)
    v = np.linspace(0, 2*np.pi, 60)
    U, V = np.meshgrid(u, v)
    X = U
    eq = abs((U**2) * (7-(U**2)))/2
    Y1 = (np.sqrt(eq)*np.cos(V))
    Z1 = (np.sqrt(eq)*np.sin(V))
    
    Y2 = (np.sqrt(eq)*np.sin(V))
    Z2 = (np.sqrt(eq)*np.cos(V))
    plt.ion()
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{P_x}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_wireframe(X, Y1, Z1, alpha=0.3, color='red')
    
    ax = fig.add_subplot(1, 3, 2, projection='3d')
    plt.ion()
    plt.axis('on')
    ax.set_box_aspect((1, 2, 1))
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{P_y}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_wireframe(Y2, X, Z2, alpha=0.3, color='blue')
    
    ax = fig.add_subplot(1, 3, 3, projection='3d')
    plt.ion()
    plt.axis('on')
    ax.set_box_aspect((1, 1, 2))
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{P_z}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_wireframe(Y2, Z2, X, alpha=0.3, color='green')
    plt.tight_layout()
    plt.show()

```
