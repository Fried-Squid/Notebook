```python
%matplotlib ipympl
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
#suppresses weird generator output
from contextlib import contextmanager
import sys, os

@contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        old_stdout = sys.stdout
        sys.stdout = devnull
        try:  
            yield
        finally:
            sys.stdout = old_stdout

with suppress_stdout():
    theta, phi = np.linspace(0, 2 * np.pi, 40), np.linspace(0, np.pi, 40)
    THETA, PHI = np.meshgrid(theta, phi)
    R =  ((np.sqrt(15)*np.sin(THETA)*np.cos(THETA)/2)*(np.sin(PHI)/np.sqrt(np.pi)))**2
    X = R * np.sin(PHI) * np.cos(THETA)
    Y = R * np.sin(PHI) * np.sin(THETA)
    Z = R * np.cos(PHI)
    fig = plt.figure()
    
    ax = fig.add_subplot(2,3,1, projection='3d')
    ax.set_box_aspect((2, 2, 0.5))
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{D_{xy}}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_surface(
        X, Y, Z, rstride=1, cstride=1,
        linewidth=0, antialiased=True, alpha=.5,color='purple')
    
    ax = fig.add_subplot(2,3,3, projection='3d')
    theta, phi = np.linspace(0, 2 * np.pi, 40), np.linspace(0, np.pi, 40)
    THETA, PHI = np.meshgrid(theta, phi)
    R =  ((np.sqrt(15)*np.sin(PHI)*np.cos(PHI)/2)*(np.sin(THETA)/np.sqrt(np.pi)))**2
    X = R * np.sin(PHI) * np.cos(THETA)
    Y = R * np.sin(PHI) * np.sin(THETA)
    Z = R * np.cos(PHI)
    ax.set_box_aspect((0.5, 2, 2))
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{D_{zy}}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_surface(
        X, Y, Z, rstride=1, cstride=1,
        linewidth=0, antialiased=True, alpha=.5,color='yellow')
    
    ax = fig.add_subplot(2,3,2, projection='3d')
    theta, phi = np.linspace(0, 2 * np.pi, 40), np.linspace(0, np.pi, 40)
    THETA, PHI = np.meshgrid(theta, phi)
    R =  ((np.sqrt(15)*np.sin(THETA)*np.cos(PHI)/2)*(np.sin(PHI)/np.sqrt(np.pi)))**2
    Y = R * np.sin(PHI) * np.cos(THETA)
    X = R * np.sin(PHI) * np.sin(THETA)
    Z = R * np.cos(PHI)
    ax.set_box_aspect((2, 0.5, 2))
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{D_{xz}}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_surface(
        X, Y, Z, rstride=1, cstride=1,
        linewidth=0, antialiased=True, alpha=.5,color='cyan')

    
    ax = fig.add_subplot(2,3,4, projection='3d')
    theta, phi = np.linspace(0, 2 * np.pi, 40), np.linspace(0, np.pi, 40)
    THETA, PHI = np.meshgrid(theta, phi)
    T_OFFSET = 0.25*np.pi
    R =  ((np.sqrt(15)*np.sin(THETA+T_OFFSET)*np.cos(THETA+T_OFFSET)/2)*(np.sin(PHI)/np.sqrt(np.pi)))**2
    X = R * np.sin(PHI) * np.cos(THETA)
    Y = R * np.sin(PHI) * np.sin(THETA)
    Z = R * np.cos(PHI)
    ax.set_box_aspect((2, 2, .5))
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{D_{x^2-y^2}}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_surface(
        X, Y, Z, rstride=1, cstride=1,
        linewidth=0, antialiased=True, alpha=.5,color='grey')
    
    ax = fig.add_subplot(2,3,5, projection='3d')
    theta, phi = np.linspace(0, 2 * np.pi, 40), np.linspace(0, np.pi, 40)
    THETA, PHI = np.meshgrid(theta, phi)
    T_OFFSET = 0 
    R = ((np.sqrt(10)*((3*(np.cos(PHI))**2)-1)/4)*(1/np.sqrt(2*np.pi)))**2
    X = R * np.sin(PHI) * np.cos(THETA)
    Y = R * np.sin(PHI) * np.sin(THETA)
    Z = R * np.cos(PHI)
    ax.set_box_aspect((1,1,2))
    plt.axis('on')
    ax.scatter([0], [0], [0], color="black", s=100)
    ax.set_title('$\mathregular{D_{z^2}}$ orbitals')
    ax.grid(True)
    ax.w_xaxis.line.set_color("red")
    ax.w_yaxis.line.set_color("blue")
    ax.w_zaxis.line.set_color("green")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.plot_surface(
        X, Y, Z, rstride=1, cstride=1,
        linewidth=0, antialiased=True, alpha=.4,color='lime')
```
