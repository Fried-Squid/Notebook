<span style="font-size: 30px">Shells.</span> 
<hr>
<ul>
    <li>
    Electrons fill successive shells, each of which is progressively further away from the nucleus.</li>
    <ol>
        <li>As distance from the nucleus increases, so does the energy level of that shell.</li>
        <li>The <b>principle quantum number</b> (n) of a shell is a measure of the energy level.</li>
        <li>The closer to the nucleus, (the lower the energy level) the lower n becomes. At the lowest energy level, n=1.</li>
    </ol> 
    <li>Each shell contains $2n^2$ electrons - the lowest energy level shell, for example, contains 2 electrons.</li>

        <li>"D" shells, observed at 3d, through to 4d and 5d, etc</li>
        <li>"F" shells, observed at 4f, through to 5f and 6f, etc</li>

        <span style="font-size: 30px">Electron Orbitals</span>
     
<hr>
<ul><li><a href="https://www.youtube.com/watch?v=Ewf7RlVNBSA">Quick orbitals explanation</a></li>
    <li>
    Subshells can be further divided, to differentiate the electrons.
    </li>
    <li>
    These sub-divisions are called "orbitals"
    </li> 
    <li>
    An orbital is not a tangible thing. An orbital is a probability cloud, where there is a 90-95% chance to find an electron.
    </li>
    <li>
    There is a low chance to find an electron outside of an orbital, and that chance decreases as the distance to the nucleus increases
    </li> 
    <li>
        <b>Heisenberg's uncertainty principle</b> states that it is impossible to know the path of an electron, only its position in space when observed.
    </li><li>Orbitals of exactly the same energy are said to be degenerate.</li>
    <li>
    There are 4 types of orbital, and some sub-types of orbitals
    </li> 
    <li>
    There are S, P, D and F orbitals, as well as $\ce{P_{x},\ P_{y},\, \ P_{z}},\ \ce{D_{xy},\ D_{yz},\ D_{zx},\ D_{x^2y^2}\ and\ D_{z^2}}$ orbitals, which will make sense once you see the graphical representation.
    </li>
    <li>
    Each subshell has different amounts of orbitals within them.
    </li> 
    <li>
        <ul><li>A S sub-shell contains 1 s-orbital.</li><li>A P sub-shell contains three P-orbitals.</li><li>A D sub-shell contains 5 D-orbitals.</li><li>An F subshell contains 7 F-orbitals.</li></ul>
    </li> 
    <li>

    <li>
    Each orbital can contain up to 2 electrons.

        <span style="font-size: 30px">Box Notation</span><hr>
        <ul> 
    <li>Atomic orbitals can be represented by boxes on an energy diagram.</li>
    <li><img src="static/Cl box notation.png"></li>
    <li>Electrons, represented by arrows, fill up these boxes to show electron configuration.</li>
    <li>The direction of the arrow represents the spin of the electron. In order to minimise repulsive forces, the spins of 2 electrons, paired in one orbital, are opposite.</li>
    <li><b>The Aufbau Principle - </b> The Aufbau Principle states that in the ground state (the lowest energy/most stable state) of an atom or ion, electrons will fill from the lowest energy orbitals first.</li><li>Ex. Lithium's electron configuration is 2 electrons in 1s, and one in 2s, rather than, say 1 in 1s, 1 in 2s and 1 in 2p.</li>
    <li><b>Hund's Rule of maximum multiplicity - </b>Hund's Rule of maximum multiplicity states that orbitals in the same sub-shell are preferentially singly occupied, meaning that electrons will only pair up if there are no more empty orbitals in a sub-shell.</li>

<span style="font-size: 30px">Bus Analogy</span><hr>
<ul>
    <li>It is easier to sit on the lower deck of a bus than the upper deck of a bus.</li>

<span style="font-size: 30px">Electron Configuration Notation</span><hr>
<ul>
    <li>Full Configuration:</li><li><ul><li>The full configuration includes all shells, sub-shells and orbtials.</li><li>The amount of electrons is represented by the power on the shell.</li><li>Ex: $1s^{2}2s^1$ is Lithium and $1s^{2}2s^{2}2p^{5}3s^2$ is cholorine</ul></li><li>Short Configuration</li><li><ul><li>Short configuration ignores the full shells to convey more useful information.</li><li>To find the short configuration, you write out the full configuration and replace the first part of the configuration with a noble gas.</li><li>Ex: $[He]2s^1$ would be lithium and $[Ne]3s^{2}3p^5$ is chlorine</ul></li>
