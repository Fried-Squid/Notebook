 <span style="font-size: 30px">Expanding and factorising</span>
<hr>
<ul>
    <li>
    A factorised expression is one which is exprressed using only products of expressions within brackets.<br>
    </li>
        <li>
    $(x+2)(-x+y)$ is factorised<br>
    </li>
        <li>
    $(x+2)(-x+y)+(19x*y)$ is not.<br>
    </li>
        <li>
    The roots of a quadratic can be used to find the factorised form.<br>
    </li>
        <li>
    $\large{x^2-5x-14 = 0 \therefore x_1=7\ \text{and}\ x_2=-2}$<br>
    </li>
    <li>
    roots can be re-written as a fraction in the form $\frac{a}{b}$: $\frac{7}{1}$ and $\frac{-2}{1}$<br>
    </li>
        <li>
   sub into $(b_1x-a_1)(b_2x-a_2)$ makes $(1x-7)(1x+2)$<br>
    </li>

    
 </ul>
